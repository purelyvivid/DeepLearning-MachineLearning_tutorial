```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
```

# Dataset

```python
dataset = load_boston()
```


    dataset['target'] - 1D numpy array of target attribute values
    dataset['data'] - 2D numpy array of attribute values
    dataset['feature_names'] - 1D numpy array of names of the attributes
    dataset['DESCR'] - text description of the dataset

```python
print(dataset.DESCR)
```

# LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
LinearRegression?
```

```python
LR = LinearRegression()
```

```python
LR.fit(dataset.data,dataset.target)
```

```python
LR.coef_
```

```python
df=pd.DataFrame(index=dataset.feature_names,data=LR.coef_,columns=['coeff'])
df['coeff_abs'] = np.abs(df['coeff'])
df.sort_values('coeff_abs',ascending=False,inplace=True)
df.head()
```

```python
LR.score(dataset.data,dataset.target) #R^2
```

```python
pred = LR.predict(dataset.data) 
```
